```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('BeijingPMtrain.csv')

data.head()
```

```python
plt.figure(figsize=(10,6))
plt.grid(linestyle='--',alpha=0.5)
plt.plot(data['hour'],data['TEMP'],data['Iws'])
idx = np.arange(0,len(data),6)
plt.xticks(idx,data['hour'][idx],rotation='vertical')
plt.xlabel('hour')
plt.ylabel('Humi,Temp')
plt.show()
```
